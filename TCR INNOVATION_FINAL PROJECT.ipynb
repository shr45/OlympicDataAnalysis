```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('athlete_events.csv')
df.head()
```

```python
df.describe()
```

```python
df["Age"].fillna((df["Age"].mean()), inplace = True)
df["Height"].fillna((df["Height"].mean()), inplace = True)
df["Weight"].fillna((df["Weight"].mean()), inplace = True)
```

```python
df[["Age", "Height", "Weight"]]= df[["Age", "Height","Weight"]].astype(int)
```

# Analysis of age,height and weight of athelete

```python
df_stat_player = df.groupby("Sport")[["Age", "Height", "Weight"]].agg([np.mean, np.max, np.min]).reset_index()
df_stat_player
```

# NUMBER OF ATHLETE

```python
total_athlete = df.groupby("Season")["Sex"].count()
print("Total number of athlete in both season\n", total_athlete)
```

```python
Total_count = df.groupby("Season")[["Team", "NOC", "Event", "City", "Sport"]].nunique().reset_index()
Total_count
```

# ATHLETE PARTICIPATION ACCORDING TO SPORT

```python
df_sports = df.groupby("Sport")["ID"].count()
print("Participation of athletes in each sports:\n", df_sports)
```

# PARTICIPATION ACCORDING TO THE YEAR IN SPORT

```python
df_year = df.groupby("Year")["Sport"].count().reset_index()
df_year.head()
```

# CREATING GOLD MEDAL DATAFRAME

```python
gold_medals = df[(df.Medal == "Gold")]
gold_medals.head()
```

```python
gold_medals = gold_medals[np.isfinite(gold_medals["Age"])]
gold_medals.head()
```

```python
gold_medals["ID"][gold_medals["Age"]> mean_age].count()
```

```python
gold_medals["ID"][gold_medals["Age"]< mean_age].count()
```

```python
event_medal_above = gold_medals["Sport"][gold_medals["Age"] > mean_age]
event_medal_above.head()
```

```python
df_M_F = df_summer.groupby(['Sex', 
'Year']).size().reset_index().pivot(columns='Sex', index='Year', values=0).reset_index()
df_M_F.head()
```
